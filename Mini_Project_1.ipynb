```python
df = df.drop_duplicates()

print(df)
```

```python
total_customer = df["Customer Id"].nunique()

print(f"Total number of customers: {total_customer}")
```

```python
product_customers = df.groupby("Company")["Customer Id"].count()

print(product_customers)
```

```python
product_customers = df.groupby("Company")["Customer Id"].count()

product_customers.plot(kind='bar', figsize=(12, 6))

plt.title("Customer Count by Company")

plt.xlabel("Company")

plt.ylabel("Number of Customers")

plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
