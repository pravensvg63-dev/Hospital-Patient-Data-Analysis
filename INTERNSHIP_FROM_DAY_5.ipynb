```python
import pandas  as pd

data ={
    'Product':['Laptop','mobile','None','Tablet','Laptop'],
    'Sales':[50000,30000,20000,None,50000],
    'Quantity':[5,3,None,2,5]
}

df = pd.DataFrame(data)

print("orginal Dataset:")
print(df)

df = df.drop_duplicates()

df['Products'] = df['Product'].fillna('unknown')

df['Sales'] = df['Sales'].fillna(df['Sales'].mean())

df['Quantity'] = df['Quantity'].fillna(0)

print("\ncleaned Dataset:")
print(df)
```

```python
region_sales = df.groupby("Product")["Sales"].sum()

print(region_sales)
```

```python
import pandas as pd

data ={
    'Product':['Laptop','mobile','None','Tablet','Laptop'],
    'Sales':[50000,30000,20000,None,50000],
    'Quantity':[5,3,None,2,5]
}

df = pd.DataFrame(data)

df = df.drop_duplicates()

df['Products'] = df['Product'].fillna('unknown')

df['Sales'] = df['Sales'].fillna(df['Sales'].mean())

df['Quantity'] = df['Quantity'].fillna(0)

Product_sales = df.groupby("Product")["Sales"].sum()

print(Product_sales)
```

```python
import matplotlib.pyplot as plt

# Line chart data from a previous cell
months = ["Jan","Feb","March","Apr","may"]
sales = [10000, 12000, 15000, 13000, 18000]

# Assuming a figure and first subplot (line chart) for context
plt.figure(figsize=(15, 5)) # Added to ensure proper subplot display

# Line chart (assuming this is the first subplot the user intended)
plt.subplot(1,3,1) # Added this subplot for consistency with 1,3,2 and 1,3,3
plt.plot(months, sales)
plt.title("Monthly Sales")
plt.xlabel("Months")
plt.ylabel("Sales Amount")

# Bar chart
plt.subplot(1,3,2) # changed to 1 row, 3 columns, position 2

plt.bar(months,
        sales, # Corrected to 'sales' (lowercase) from original definition
        color="orange")

plt.title("Sales Comparison")

plt.xlabel("Months")
plt.ylabel("Sales")

#Pie chart
plt.subplot(1,3,3) # changed to 1 row, 3 columns, position 3

expenses = [5000,3000,2000]
categories = ["Food","travel","Shopping"]
plt.pie(expenses,
        labels=categories,
        autopct="%1.1f%%")

plt.title("Expenses Distribution")

plt.show()
```
